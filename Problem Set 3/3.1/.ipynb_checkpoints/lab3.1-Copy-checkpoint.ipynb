### (a)

Generate term x color-chip matrix of counts from `data/term.txt`.

Note: in `data/term.txt`, this is what the indices represent:
 * 0: WCS Language Number 
 * 1: WCS Speaker Number
 * 2: WCS Chip Number
 * 3: Term Abbreviation
 
Skip rows that start have '\*' as the term abbreviation (represents no answer).

Note that the term abbreviations are per-language and are not unique. This means that for each term, you need to keep track of (language_number, term_abbreviation) pairs.

```python
from collections import defaultdict, Counter
import numpy as np
from tqdm import tqdm


def generate_color_chip_matrix(output_file, input_file, lang_filter=None):
    chip_to_term_count = defaultdict(Counter)
    unique_terms = set()
    with open(input_file) as f:
        for line in tqdm(f):
            lang_num, speaker_num, chip_num, term_abbr = line.split()
            if lang_filter and lang_num not in lang_filter:
                continue
            if term_abbr == '*':
                continue
            unique_term_label = lang_num + '-' + term_abbr
            unique_terms.add(unique_term_label)
            chip_to_term_count[int(chip_num)].update([unique_term_label])
    
    term_to_idx = {unique_term: i for i, unique_term in enumerate(unique_terms)}
    assert len(chip_to_term_count) == 330
    result = np.zeros((len(unique_terms), len(chip_to_term_count)))
    for chip_num, term_counter in chip_to_term_count.items():
        for curr_term, curr_count in term_counter.items():
            result[term_to_idx[curr_term], chip_num - 1] = curr_count
    row_labels = [None for __ in range(len(term_to_idx))]
    for curr_term, curr_index in term_to_idx.items():
        row_labels[curr_index] = curr_term
    write_matrix(output_file, result, row_labels)


def write_matrix(output_file, matrix, row_labels):
    """
    Each line is tab-separated. The first item is the row_label identifier.
    The remaining items are for the 330 Munsell color chips in order.
    """
    with open(output_file, 'w') as f:
        for i, row in enumerate(matrix):
            f.write(row_labels[i] + '\t' + '\t'.join([str(item) for item in row.tolist()]) + '\n')
    

def load_matrix(matrix_file):
    row_labels = []
    row_label_to_row = {}
    with open(matrix_file, 'r') as f:
        for line in f:
            line = line.strip().split('\t')
            row_labels.append(line[0])
            row_label_to_row[line[0]] = np.array([float(item) for item in line[1:]])
    result = np.zeros((len(row_labels), len(row_label_to_row[row_labels[0]])))
    for i, row_label in enumerate(row_labels):
        result[i] = row_label_to_row[row_label]
    return result, row_labels
    

matrix_file = 'data/color_chip_matrix.tsv'
generate_color_chip_matrix(matrix_file, 'data/term.txt')
```

### (c)

Write a function to compute a distance matrix of color x color, where each cell is the distance between two colors using euclidean distance. Call this function on the matrix from part b.

```python
from scipy.spatial.distance import euclidean


def generate_distance_matrix(matrix):
    n_chips = matrix.shape[1]
    dist_matrix = np.zeros((n_chips, n_chips))
    for i in range(n_chips):
        for j in range(i + 1, n_chips):
            result = euclidean(matrix[:, i], matrix[:, j])
            dist_matrix[i, j] = dist_matrix[j, i] = result

    # normalize
    for i in range(len(dist_matrix)):
        dist_matrix[i, :] = dist_matrix[i, :] / np.sum(dist_matrix[i, :])
    return dist_matrix


matrix_pca_file = 'data/color_chip_matrix.tsv'
color_chip_pca_mat, term_labels = load_matrix(matrix_pca_file)
chip_distance_matrix = generate_distance_matrix(color_chip_pca_mat)
```

### (d)

Generate a matrix of L\*a\*b\* by color. Call the function from part c on this matrix.

```python
def generate_perceptual_color_matrix(output_file, input_file):
    result = np.zeros((3, 330))
    with open(input_file) as f:
        next(f)
        for line in f:
            line = line.strip().split()
            chip_num = int(line[0])
            lab_values = [float(item) for item in line[6:]]
            result[:, chip_num - 1] = lab_values        
    write_matrix(output_file, result, ['L*', 'a*', 'b*'])

generate_perceptual_color_matrix('data/perceptual_color_mat.tsv', 'data/cnum-vhcm-lab-new.txt')
```

```python
percept_color_mat_file = 'data/perceptual_color_mat.tsv'
perceptual_color_mat, __ = load_matrix(percept_color_mat_file)
perceptual_distance_matrix = generate_distance_matrix(perceptual_color_mat)
```

### (e)

Compute the r squared score of the distance matrices computed in part c (cross-linguistic space) and part d (perceptual space). Repeat the computation for copies of the matrices from part c and d where the rows are randomly scrambled. How do these two values compare? (We expect it to be more similar for the unscrambled matrices)

```python
from sklearn.metrics import r2_score
import random


def shuffle_matrix_by_row(matrix):
    result = matrix.copy()
    for i in range(len(result)):
        random.shuffle(result[i, :])
    return result


print("r2_score {}".format(r2_score(perceptual_distance_matrix, chip_distance_matrix)))
print("r2_score for both shuffled {}".format(r2_score(
    shuffle_matrix_by_row(perceptual_distance_matrix),
    shuffle_matrix_by_row(chip_distance_matrix))))
print("r2_score for one shuffled {}".format(r2_score(
    perceptual_distance_matrix,
    shuffle_matrix_by_row(chip_distance_matrix))))
```

### (f)

Compare the r squared scores with perceptual space for subsets of languages -- is the correlation higher for Western European languages than for African languages, for example?

I think that my main take-aways are:
 * None of the language families correlate as well with the perceptual space as the cross-linguistic distance matrix with the full set of languages does. This is probably because the different language families encode different contrasts within the perceptual space. My thinking is: the more varied the sample, the more contrasts that are represented, the higher the correlation with the perceptual space.
 * Some language families, like the Niger-Congo language family do particularly badly. I thought this might have to do with sparsity of data, but in fact the Niger-Congo language family is the language family with the highest number of languages in the sample. Perhaps the correlation has more to do with how much variation exists within a language family.

```python
from collections import defaultdict
from sklearn.metrics import r2_score
from sklearn.decomposition import PCA


def get_language_families(min_members=5):
    result = defaultdict(list)
    with open('data/language_families.tsv') as f:
        next(f)  # skip headers
        for line in f:
            line = line.strip().split('\t')
            lang_id = line[0]
            lang_fam = line[-1]
            if lang_fam == 'Isolate':
                continue
            result[lang_fam].append(lang_id)
    
    result_filtered = {}
    for fam, langs in result.items():
        if len(langs) >= min_members:
            result_filtered[fam] = langs
    return result_filtered


def evaluate_language_family_correlations(gold_standard_distance_matrix):
    family_to_vals = get_language_families()
    for family, family_filter in family_to_vals.items():
        print('{}:'.format(family))
        
        # generate color chip mat for this language family
        output_file = 'data/chip_matrix_{}.tsv'.format(family)
        generate_color_chip_matrix(output_file, 'data/term.txt', family_filter)
        color_chip_mat, term_labels = load_matrix(output_file)
        
        # call PCA on the color chip mat for this language famliy
        pca = PCA(n_components=min(100, color_chip_mat.shape[0] // 2))
        color_chip_pca_mat = pca.fit_transform(color_chip_mat.T).T
        
        # get the distance matrix based on the PCA matrix
        chip_distance_matrix = generate_distance_matrix(color_chip_pca_mat)
        
        # compute the r squared value relative to gold_standard_distance_matrix
        print("\tr2_score {}".format(r2_score(gold_standard_distance_matrix, chip_distance_matrix)))
        print("\tr2_score for shuffled {}".format(r2_score(
            shuffle_matrix_by_row(gold_standard_distance_matrix),
            shuffle_matrix_by_row(chip_distance_matrix))))


percept_color_mat_file = 'data/perceptual_color_mat.tsv'
perceptual_color_mat, __ = load_matrix(percept_color_mat_file)
perceptual_distance_matrix = generate_distance_matrix(perceptual_color_mat)
evaluate_language_family_correlations(perceptual_distance_matrix)
```

### (g)

Create visualisations for the perceptual and cross linguistic dimensions using pca with n_components = 2. Can you tell that the cross-linguistic space is warped compared to the perceptual space?

Note: for the cross-linguistic space, do the PCA on the original matrix, rather than on the pca-ed matrix with dim=100.

```python
from colormath.color_objects import LabColor, sRGBColor
from colormath.color_conversions import convert_color
import matplotlib.pyplot as plt
import numpy as np


PERCEPTUAL_MATRIX_FILE = 'data/perceptual_color_mat.tsv'


def get_perceptual_matrix_rgb():
    perceptual_mat, __ = load_matrix(PERCEPTUAL_MATRIX_FILE)
    perceptual_mat = perceptual_mat.T
    result_mat = np.zeros(perceptual_mat.shape)
    for i, row in enumerate(perceptual_mat):
        l, a, b = row
        lab = LabColor(l, a, b)
        rgb = convert_color(lab, sRGBColor)
        result_mat[i, :] = [min(val, 1.0) for val in rgb.get_value_tuple()]
    return result_mat


def create_pca_graph(matrix):
    pca = PCA(n_components=2)
    color_chip_mat_pca = pca.fit_transform(matrix.T)
    x_values = [item[0] for item in color_chip_mat_pca]
    y_values = [item[1] for item in color_chip_mat_pca]
    perceptual_mat_rgb = get_perceptual_matrix_rgb()
    plt.scatter(x_values, y_values, color=[perceptual_mat_rgb[i] for i in range(len(x_values))])
    plt.show()
    
    
matrix_file = 'data/color_chip_matrix.tsv'
color_chip_mat, term_labels = load_matrix(matrix_file)
create_pca_graph(color_chip_mat)
```

```python
matrix_file = 'data/perceptual_color_mat.tsv'
perceptual_mat, term_labels = load_matrix(matrix_file)
create_pca_graph(perceptual_mat)
```

### (h)

Do clustering on the cross-linguistic space and compare that to proposed evolutionary sequences.

Note: the way that I computed the color values for the links was using the average of the two colors of the links that were merged. This is different from computing the average color across all cluster members.

Observations:
 * Stage 1: The two biggest clusters are W/R/Y and BK/G/Bu, which mirrors the evolutionary predictions.
 * Stage 2: The next division splits W/R/Y into W/Y and R. This differs from the R/Y and W in the evolutionary diagram.
 * Stage 3: The next split is G/Bu, BK, W/Y and R. This is similar to the middle option but with the same difference for the W/R/Y division as noted above.
 * Stage 4: The next division splits W/Y into W and Y. This mirrors the expected trend, leaving us with the groups W, R, Y, G/Bu, BK (the middle option).
 * Stage 5: The next split is between light blue and other white-like colors. This is followed by a split within the black colors, followed finally by the blue-green predicted split.

![Alt text](https://notebooks.azure.com/cog403/libraries/in-progress-3-1/raw/color_evolution_Kay_Cook_2009.png)

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt


def get_average_color_string(c1, c2):
    """
    adapted from:
    https://stackoverflow.com/questions/3380726/converting-a-rgb-color-tuple-to-a-six-digit-code-in-python
    """
    if isinstance(c1, str):
        c1 = [int(c1[1:3],16) / 255.0, int(c1[3:5],16) / 255.0, int(c1[5:7],16) / 255.0]
    if isinstance(c2, str):
        c2 = [int(c2[1:3],16) / 255.0, int(c2[3:5],16) / 255.0, int(c2[5:7],16) / 255.0]
    avg_color_tuple = np.mean(np.stack((c1, c2)), axis=0)
    avg_color_tuple = [255 * x for x in avg_color_tuple]
    
    #Ensure values are rounded integers, convert to hex, and concatenate
    return '#' + ''.join(['{:02X}'.format(int(round(x))) for x in avg_color_tuple])


def plot_dendrogram(dist_color_matrix):
    """
    Adapted from:
    https://stackoverflow.com/questions/38153829/custom-cluster-colors-of-scipy-dendrogram-in-python-link-color-func
    """
    Z = hierarchy.linkage(dist_color_matrix, 'complete')
   # get_cluster_classes
    plt.figure(figsize=(18, 16))
    
    leaf_colors = get_perceptual_matrix_rgb()
        
    link_cols = {}
    for i, i12 in enumerate(Z[:,:2].astype(int)):
        c1, c2 = (link_cols[x] if x > len(Z) else leaf_colors[x] for x in i12)
        link_cols[i + 1 + len(Z)] = get_average_color_string(c1, c2)
    
    dn = hierarchy.dendrogram(Z, color_threshold=None, link_color_func=lambda x: link_cols[x])
    plt.show()
    

matrix_file = 'data/color_chip_matrix_pca.tsv'
color_chip_pca_mat, __ = load_matrix(matrix_file)
chip_distance_matrix = generate_distance_matrix(color_chip_pca_mat)
plot_dendrogram(chip_distance_matrix)
```

```python
matrix_file = 'data/perceptual_color_mat.tsv'
perc_pca_mat, __ = load_matrix(matrix_file)
chip_distance_matrix = generate_distance_matrix(perc_pca_mat)
plot_dendrogram(chip_distance_matrix)
```
