### (a)  Create color-chip x term matrix encoding language data

Generate term x color-chip matrix of counts from `data/term.txt`.

**@Julia -- We've always done a situation x term matrix -- ie, in this case each color chip would be a row, and the terms from all the languages would be the columns.  Is there a particular reason you did it this way, or can we flip it?  I noticed you have to transpose it for PCA anyway.**

**SS: Call these chip_by_term_matrix_wcs and chip_by_term_matrix_bk? And similarly for function.**


Note: in `data/term.txt`, this is what the indices represent:
 * 0: WCS Language Number 
 * 1: WCS Speaker Number
 * 2: WCS Chip Number
 * 3: Term Abbreviation
 
Skip rows that have '\*' as the term abbreviation; this represents that the speaker did not give an answer.

Note that the term abbreviations are per-language and are not unique. This means that for each term, you need to keep track of (language_number, term_abbreviation) pairs. An easy way to do this is to concatenate the Language Number with the Term Abbreviation.

**@Julia -- is the idea to give them the write and load matrix functions?**

**@Suzanne -- yes, I moved them to provided_functions file**

**@Julia -- we need to think of something they can print that we can check for correctness.**

```python
from collections import defaultdict, Counter
import numpy as np
from tqdm import tqdm

from provided_functions import write_matrix, load_matrix

N_CHIPS = 330


def generate_chip_by_term_matrix(output_file, input_file, lang_filter=None):
    chip_to_term_count = defaultdict(Counter)
    unique_terms = set()
    with open(input_file) as f:
        for line in tqdm(f):
            lang_num, speaker_num, chip_num, term_abbr = line.split()
            if lang_filter and lang_num not in lang_filter:
                continue
            if term_abbr == '*':
                continue
            unique_term_label = lang_num + '-' + term_abbr
            unique_terms.add(unique_term_label)
            chip_to_term_count[int(chip_num)].update([unique_term_label])
    
    term_to_idx = {unique_term: i for i, unique_term in enumerate(unique_terms)}
    result = np.zeros((N_CHIPS, len(unique_terms)))
    for chip_num, term_counter in chip_to_term_count.items():
        for curr_term, curr_count in term_counter.items():
            result[chip_num - 1, term_to_idx[curr_term]] = curr_count
    row_labels = [None for __ in range(len(term_to_idx))]
    for curr_term, curr_index in term_to_idx.items():
        row_labels[curr_index] = curr_term
    write_matrix(output_file, result, row_labels)
    

matrix_file = 'data/chip_by_term_matrix_wcs.tsv'
generate_chip_by_term_matrix(matrix_file, 'data/term.txt')
```

### (c)  Create a color-chip x color-chip distance matrix

Write a function to compute a distance matrix of color x color, where each cell is the distance between two colors using euclidean distance. Call this function on the matrix from part b.

**@Julia -- I'm guessing we need to elaborate here.  The idea is to, given any matrix where the rows are the vector representation of the color chips, that they find the distance between all pairs of rows?  (I guess, currently it's columns where I'm saying rows.)**

**@Julia -- did you make this a full matrix, or triangular?**

```python
from scipy.spatial.distance import euclidean
import numpy as np


def generate_distance_matrix(matrix):
    n_chips = matrix.shape[0]
    dist_matrix = np.zeros((n_chips, n_chips))
    for i in range(n_chips):
        for j in range(i + 1, n_chips):
            result = euclidean(matrix[i], matrix[j])
            dist_matrix[i, j] = dist_matrix[j, i] = result

    # normalize
    for i in range(len(dist_matrix)):
        dist_matrix[i, :] = dist_matrix[i, :] / np.sum(dist_matrix[i, :])
    return dist_matrix


matrix_pca_file = 'data/chip_by_term_matrix_pca.tsv'
chip_by_term_pca_mat, term_labels = load_matrix(matrix_pca_file)
chip_distance_matrix = generate_distance_matrix(chip_by_term_pca_mat)
```

```python
def generate_perceptual_color_matrix(output_file, input_file):
    result = np.zeros((330, 3))
    with open(input_file) as f:
        next(f)
        for line in f:
            line = line.strip().split()
            chip_num = int(line[0])
            lab_values = [float(item) for item in line[6:]]
            result[chip_num - 1, :] = lab_values        
    write_matrix(output_file, result, ['L*', 'a*', 'b*'])

generate_perceptual_color_matrix('data/perceptual_color_mat.tsv', 'data/cnum-vhcm-lab-new.txt')
```

```python
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
import random


def shuffle_matrix_by_row(matrix):
    result = matrix.copy()
    for i in range(len(result)):
        np.random.shuffle(result[i, :])
    return result

print("FOR WCS DATA")

print("\nr2_score {}".format(r2_score(perceptual_distance_matrix, chip_distance_matrix)))
print("r2_score for both shuffled {}".format(r2_score(
    shuffle_matrix_by_row(perceptual_distance_matrix),
    shuffle_matrix_by_row(chip_distance_matrix))))
print("r2_score for first shuffled {}".format(r2_score(
    shuffle_matrix_by_row(perceptual_distance_matrix),
    chip_distance_matrix)))
print("r2_score for second shuffled {}".format(r2_score(
    perceptual_distance_matrix,
    shuffle_matrix_by_row(chip_distance_matrix))))

print("\npearson {}".format(pearsonr(perceptual_distance_matrix.flatten(), chip_distance_matrix.flatten())))
print("pearson for both shuffled {}".format(pearsonr(
    shuffle_matrix_by_row(perceptual_distance_matrix).flatten(),
    shuffle_matrix_by_row(chip_distance_matrix).flatten())))
print("pearson for first shuffled {}".format(pearsonr(
    shuffle_matrix_by_row(perceptual_distance_matrix).flatten(),
    chip_distance_matrix.flatten())))
print("pearson for second shuffled {}".format(pearsonr(
    perceptual_distance_matrix.flatten(),
    shuffle_matrix_by_row(chip_distance_matrix).flatten())))


print("\n\nFOR BK DATA")

print("\nr2_score {}".format(r2_score(perceptual_distance_matrix, chip_distance_matrix_bk)))
print("r2_score for both shuffled {}".format(r2_score(
    shuffle_matrix_by_row(perceptual_distance_matrix),
    shuffle_matrix_by_row(chip_distance_matrix_bk))))
print("r2_score for first shuffled {}".format(r2_score(
    shuffle_matrix_by_row(perceptual_distance_matrix),
    chip_distance_matrix_bk)))
print("r2_score for second shuffled {}".format(r2_score(
    perceptual_distance_matrix,
    shuffle_matrix_by_row(chip_distance_matrix_bk))))

print("\npearson {}".format(pearsonr(perceptual_distance_matrix.flatten(), chip_distance_matrix_bk.flatten())))
print("pearson for both shuffled {}".format(pearsonr(
    shuffle_matrix_by_row(perceptual_distance_matrix).flatten(),
    shuffle_matrix_by_row(chip_distance_matrix_bk).flatten())))
print("pearson for first shuffled {}".format(pearsonr(
    shuffle_matrix_by_row(perceptual_distance_matrix).flatten(),
    chip_distance_matrix_bk.flatten())))
print("pearson for second shuffled {}".format(pearsonr(
    perceptual_distance_matrix.flatten(),
    shuffle_matrix_by_row(chip_distance_matrix_bk).flatten())))
```

```python
from collections import defaultdict
from sklearn.metrics import r2_score
from sklearn.decomposition import PCA


def get_language_families(family_file, min_members=5):
    result = defaultdict(list)
    with open(family_file) as f:
        next(f)  # skip headers
        for line in f:
            line = line.strip().split('\t')
            lang_id = line[0]
            lang_fam = line[-1]
            if lang_fam == 'Isolate':
                continue
            result[lang_fam].append(lang_id)
    
    result_filtered = {}
    for fam, langs in result.items():
        if len(langs) >= min_members:
            result_filtered[fam] = langs
    return result_filtered


def evaluate_language_family_correlations(
    gold_standard_distance_matrix, identifier, term_file, family_file, n_trials=1000, min_members=5):
    family_to_vals = get_language_families(family_file, min_members=min_members)
    for family, family_filter in family_to_vals.items():
        print('{}:'.format(family))
        
        # generate color chip mat for this language family
        output_file = 'data/chip_matrix_{}_{}.tsv'.format(family, identifier)
        generate_chip_by_term_matrix(output_file, term_file, family_filter)
        chip_by_term_mat, term_labels = load_matrix(output_file)
        
        # call PCA on the color chip mat for this language famliy
        pca = PCA(n_components=min(100, chip_by_term_mat.shape[0] // 2))
        chip_by_term_pca_mat = pca.fit_transform(chip_by_term_mat.T).T
        
        # get the distance matrix based on the PCA matrix
        chip_distance_matrix = generate_distance_matrix(chip_by_term_pca_mat)
        
        # compute the r squared value relative to gold_standard_distance_matrix
        print("\tpearson: {}".format(pearsonr(
            gold_standard_distance_matrix.flatten(), chip_distance_matrix.flatten())))
        shuffled_pearson = []
        shuffled_significance = []
        for i in range(1000):
            result = pearsonr(
                gold_standard_distance_matrix.flatten(),
                shuffle_matrix_by_row(chip_distance_matrix).flatten())
            shuffled_pearson.append(result[0])
            shuffled_significance.append(result[1])
        print("\tpearson shuffled: {}".format((np.mean(shuffled_pearson), np.mean(shuffled_significance))))
        


percept_color_mat_file = 'data/perceptual_color_mat.tsv'
perceptual_color_mat, __ = load_matrix(percept_color_mat_file)
perceptual_distance_matrix = generate_distance_matrix(perceptual_color_mat)
evaluate_language_family_correlations(
    perceptual_distance_matrix, "wcs", 'data/term.txt', 'data/language_families.tsv')
```

### (g)  Examine the first few principal components from PCA

Create visualisations for the perceptual and cross linguistic dimensions using pca with n_components = 2. Can you tell that the cross-linguistic space is warped compared to the perceptual space?

Note: for the cross-linguistic space, do the PCA on the original matrix, rather than on the pca-ed matrix with dim=100.

**@Julia -- How can we see which colors are the ones at the "extremes" of the graphs?  Eg, the points of the triangle in the 2D projection for WCS and BK, and the 4 "points" in the 3D projections?  I'm trying to figure out how we can get them to relate this to the evolutionary diagram without just "eyeballing" it like I've been doing. :-)**

```python
from colormath.color_objects import LabColor, sRGBColor
from colormath.color_conversions import convert_color
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


PERCEPTUAL_MATRIX_FILE = 'data/perceptual_color_mat.tsv'


def get_perceptual_matrix_rgb():
    perceptual_mat, __ = load_matrix(PERCEPTUAL_MATRIX_FILE)
    perceptual_mat = perceptual_mat.T
    result_mat = np.zeros(perceptual_mat.shape)
    for i, row in enumerate(perceptual_mat):
        l, a, b = row
        lab = LabColor(l, a, b)
        rgb = convert_color(lab, sRGBColor)
        result_mat[i, :] = [min(val, 1.0) for val in rgb.get_value_tuple()]
    return result_mat


def create_pca_graph(matrix, dim=2, title=None):
    perceptual_mat_rgb = get_perceptual_matrix_rgb()
    
    pca = PCA(n_components=dim)
    chip_by_term_mat_pca = pca.fit_transform(matrix.T)
    x_values = [item[0] for item in chip_by_term_mat_pca]
    color = [perceptual_mat_rgb[i] for i in range(len(x_values))]
    
    fig = plt.figure()
    if dim == 1:
        plt.scatter(x_values, np.zeros(len(x_values)), color=color)
    if dim == 2:
        y_values = [item[1] for item in chip_by_term_mat_pca]
        plt.scatter(x_values, y_values, color=color)
    elif dim == 3:
        y_values = [item[1] for item in chip_by_term_mat_pca]
        z_values = [item[2] for item in chip_by_term_mat_pca]
        ax = Axes3D(fig)
        ax.scatter(x_values, y_values, z_values, color=color)
        
    if title:
        plt.title(title)
    plt.show()
```

```python
matrix_file = 'data/chip_by_term_matrix_wcs.tsv'
chip_by_term_mat, term_labels = load_matrix(matrix_file)
create_pca_graph(chip_by_term_mat, dim=1, title="Cross-Linguistic 1 Dimensional")
create_pca_graph(chip_by_term_mat, dim=2, title="Cross-Linguistic 2 Dimensional")
create_pca_graph(chip_by_term_mat, dim=3, title="Cross-Linguistic 3 Dimensional")
```

```python
matrix_file = 'data/perceptual_color_mat.tsv'
perceptual_mat, term_labels = load_matrix(matrix_file)
create_pca_graph(perceptual_mat, dim=1, title="Perceptual 1 Dimensional")
create_pca_graph(perceptual_mat, dim=2, title="Perceptual 2 Dimensional")
create_pca_graph(perceptual_mat, dim=3, title="Perceptual 3 Dimensional")
```

Do clustering on the cross-linguistic space and compare that to proposed evolutionary sequences.

Note: the way that I computed the color values for the links was using the average of the two colors of the links that were merged. This is different from computing the average color across all cluster members.

Observations:
 * Stage 1: The two biggest clusters are W/R/Y and BK/G/Bu, which mirrors the evolutionary predictions.
 * Stage 2: The next division splits W/R/Y into W/Y and R. This differs from the R/Y and W in the evolutionary diagram.
 * Stage 3: The next split is G/Bu, BK, W/Y and R. This is similar to the middle option but with the same difference for the W/R/Y division as noted above.
 * Stage 4: The next division splits W/Y into W and Y. This mirrors the expected trend, leaving us with the groups W, R, Y, G/Bu, BK (the middle option).
 * Stage 5: The next split is between light blue and other white-like colors. This is followed by a split within the black colors, followed finally by the blue-green predicted split.

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt


def get_average_color_string(c1, c2):
    """
    adapted from:
    https://stackoverflow.com/questions/3380726/converting-a-rgb-color-tuple-to-a-six-digit-code-in-python
    """
    if isinstance(c1, str):
        c1 = [int(c1[1:3],16) / 255.0, int(c1[3:5],16) / 255.0, int(c1[5:7],16) / 255.0]
    if isinstance(c2, str):
        c2 = [int(c2[1:3],16) / 255.0, int(c2[3:5],16) / 255.0, int(c2[5:7],16) / 255.0]
    avg_color_tuple = np.mean(np.stack((c1, c2)), axis=0)
    avg_color_tuple = [255 * x for x in avg_color_tuple]
    
    #Ensure values are rounded integers, convert to hex, and concatenate
    return '#' + ''.join(['{:02X}'.format(int(round(x))) for x in avg_color_tuple])


def plot_dendrogram(dist_color_matrix):
    """
    Adapted from:
    https://stackoverflow.com/questions/38153829/custom-cluster-colors-of-scipy-dendrogram-in-python-link-color-func
    """
    Z = hierarchy.linkage(dist_color_matrix, 'complete')
   # get_cluster_classes
    plt.figure(figsize=(18, 16))
    
    leaf_colors = get_perceptual_matrix_rgb()
        
    link_cols = {}
    for i, i12 in enumerate(Z[:,:2].astype(int)):
        c1, c2 = (link_cols[x] if x > len(Z) else leaf_colors[x] for x in i12)
        link_cols[i + 1 + len(Z)] = get_average_color_string(c1, c2)
    
    dn = hierarchy.dendrogram(Z, color_threshold=None, link_color_func=lambda x: link_cols[x])
    plt.show()
    

matrix_file = 'data/chip_by_term_matrix_pca.tsv'
chip_by_term_pca_mat, __ = load_matrix(matrix_file)
chip_distance_matrix = generate_distance_matrix(chip_by_term_pca_mat)
plot_dendrogram(chip_distance_matrix)
```
