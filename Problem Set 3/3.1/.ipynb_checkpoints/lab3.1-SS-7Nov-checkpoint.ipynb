Access your matrix to print how many times the PR label for the 49th language in the WCS data is used to describe color chip 7.

```python
from collections import defaultdict, Counter
import numpy as np
from tqdm import tqdm

from provided_functions import write_matrix, load_matrix

N_CHIPS = 330


def normalize(matrix):
    for i in range(len(matrix)):
        if np.sum(matrix[i, :]) != 0:
            matrix[i, :] = matrix[i, :] / np.sum(matrix[i, :])
    return matrix


def generate_chip_by_term_matrix(output_file, input_file, lang_filter=None):
    chip_to_term_count = defaultdict(Counter)
    unique_terms = set()
    with open(input_file) as f:
        for line in tqdm(f):
            lang_num, speaker_num, chip_num, term_abbr = line.split()
            if lang_filter and lang_num not in lang_filter:
                continue
            if term_abbr == '*':
                continue
            unique_term_label = lang_num + '-' + term_abbr
            unique_terms.add(unique_term_label)
            chip_to_term_count[int(chip_num)].update([unique_term_label])
    
    term_to_idx = {unique_term: i for i, unique_term in enumerate(unique_terms)}
    result = np.zeros((N_CHIPS, len(unique_terms)))
    for chip_num, term_counter in chip_to_term_count.items():
        for curr_term, curr_count in term_counter.items():
            result[chip_num - 1, term_to_idx[curr_term]] = curr_count
    row_labels = [None for __ in range(len(term_to_idx))]
    for curr_term, curr_index in term_to_idx.items():
        row_labels[curr_index] = curr_term
    
    # I added this code to normalize the matrix
    result = normalize(result)
    
    write_matrix(output_file, result, row_labels)
    

matrix_file = 'data/chip_by_term_matrix_wcs.tsv'
generate_chip_by_term_matrix(matrix_file, 'data/term.txt')

# print something we can check for correctness
chip_by_term_wcs, term_labels = load_matrix(matrix_file)
print(chip_by_term_wcs[7 - 1, term_labels.index('49-PR')])  # subtract one, since it's zero-indexed
```

```python
from scipy.spatial.distance import euclidean
import numpy as np


def generate_distance_matrix(matrix):
    n_chips = matrix.shape[0]
    dist_matrix = np.zeros((n_chips, n_chips))
    for i in range(n_chips):
        for j in range(i + 1, n_chips):
            result = euclidean(matrix[i], matrix[j])
            dist_matrix[i, j] = dist_matrix[j, i] = result
    return dist_matrix


test_case = np.array([[1, 1],
                      [1, 2],
                      [4, 5]])
print(generate_distance_matrix(test_case))
```

###  Calculate a distance matrix in the perceptual space

The L\*a\*b\* space is a color space with three dimensions: L\* represents lightness, and a\* represents the spectrum from green to red, and b\* represents the spectrum from blue to yellow. The most cognitively interesting feature of this space is that it was designed to be perceptually uniform, meaning that the distance between two points in the L\*a\*b\* space is correlated with humans percieved diffences (rather than with physical properties of the colors). For more information on the L\*a\*b\* space, see this [link](https://en.wikipedia.org/wiki/CIELAB_color_space).

The L\*a\*b\* values of the color chips are stored in `data/perceptual_color_mat.tsv`.

Create visualisations for the perceptual and cross linguistic dimensions using pca with n_components = 2. Can you tell that the cross-linguistic space is warped compared to the perceptual space?

```python
from collections import Counter
import numpy as np
from scipy.spatial.distance import euclidean
from sklearn.neighbors import NearestNeighbors

from provided_functions import load_matrix


def get_english_chip_labels():
    chip_to_label = {}
    with open('data/chip_en_labels.txt', 'r') as f:
        for line in f:
            line = line.strip('\n').split('\t')
            chip_to_label[int(line[0])] = line[-1]
    return [chip_to_label[i + 1] for i in range(len(chip_to_label))]
            
        

def identify_extreme_points(matrix, n_neighbors=10):
    """
    matrix: a 2D np.array
    """
    extreme_points = set()
    for i in range(matrix.shape[1]):
        min_extreme_point = np.argmin(matrix[:, i])
        max_extreme_point = np.argmax(matrix[:, i])
        extreme_points.add(min_extreme_point)
        extreme_points.add(max_extreme_point)
            
    # heuristic: remove cases where the distance between two points
    # is less than distance of either point the mean point
    # (avoids having two extreme points very close to each other)
    extreme_points = list(extreme_points)
    extreme_values = [tuple(matrix[i]) for i in extreme_points]
    mean_point = np.mean(matrix, axis=0)
    distances_from_mean = [euclidean(mean_point, x) for x in extreme_values]
    points_to_remove = set()
    for i in range(len(extreme_values)):
        point1 = extreme_values[i]
        for j in range(i + 1, len(extreme_values)):
            point2 = extreme_values[j]
            if euclidean(point1, point2) < min(
                    distances_from_mean[i], distances_from_mean[j]):
                if euclidean(point1, np.zeros(len(point2))) > euclidean(point2, np.zeros(len(point1))):
                    points_to_remove.add(point2)
                else:
                    points_to_remove.add(point1)
    for point in points_to_remove:
        index = extreme_values.index(point)
        extreme_points.pop(index)
        extreme_values.pop(index)
    
    # find nearest neighbors of extreme points and use them to determine colors
    nbrs = NearestNeighbors(n_neighbors=n_neighbors, algorithm='ball_tree').fit(matrix)
    english_labels = get_english_chip_labels()
    for index, point in zip(extreme_points, extreme_values):
        distances, indices = nbrs.kneighbors([point])
        indices = indices[0]
        # no need to add index back in because it was found
        color_label_counter = Counter([english_labels[i] for i in indices])
        best_label = color_label_counter.most_common(1)[0]
        print("{} ({:.2f}%): {}".format(best_label[0], (best_label[1] / n_neighbors) * 100, point))
```
